리스트를 만들 때 필요한 함수 : append()


```python
scores=[]
while True:
  score=input("학생의 점수를 입력하세요:")
  if score.isdigit():
    scores.append(int(score))
  else:
    break
print(scores)
print("합계:", sum(scores))
print("인원수:", len(scores))
def average(x,y):
  s = x/y
  return s
print("평균:", average(sum(scores),len(scores)) )
```

```python
scores = [88, 92, 78, 60, 75, 95]
def convert_to_grade(score):
 if score >= 90:
    return 'A'
 elif score >= 80:
    return 'B'
 elif score >= 70:
    return 'C'
 elif score >= 60:
    return 'D'
 else:
    return 'F'
print(list(map(convert_to_grade,scores)))
```
