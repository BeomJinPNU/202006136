```python
fruitsnumber = int(input("과일 개수를 입력하세요: "))
fruitslist = {}
for _ in range(fruitsnumber):
    fruit, price = input("과일명과 가격을 입력하세요: ").split()
    fruitslist[fruit] = int(price)

the_number_of_fruits = int(input("구매한 과일의 종류가 몇 개인지 입력하세요: "))
purchaselist = {}
for _ in range(the_number_of_fruits):
    fruit, quantity = input("구매한 과일명과 개수를 입력하세요: ").split()
    purchaselist[fruit] = int(quantity)

def calculate_total_price(fruits_prices, purchase):
    total_price = 0
    for fruit, quantity in purchase.items():
        if fruit in fruits_prices:
            total_price += fruits_prices[fruit] * quantity
        else:
            break
    return total_price

total_price = calculate_total_price(fruitslist, purchaselist)
print(total_price)
```
