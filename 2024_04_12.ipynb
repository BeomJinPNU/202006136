```python
def square(x):
    return x ** 2

def square_list(numbers):
    squared_numbers = list(map(square, numbers))
    return squared_numbers

# 예시 리스트
my_list = [1, 2, 3, 4, 5]

# 함수 호출하여 결과 출력
print(square_list(my_list))
```
