```python
#데이터프레임 생성
df=pd.DataFrame({'학생': ['John','Alice','Bob','Emily'],
                 '수학': [90, 87, 78, 92],
                 '영어': [85, 91, 80, 88],
                 '과학': [88, 89, 85, 82]})
print(df)
#요약 통계 정보 출력
print(df.describe())
#'수학'점수가 가장 높은 학생의 정보를 출력
print(df[df['수학']==df['수학'].max()])
#'수학'점수가 90점 이상인 학생의 정보를 출력
morethan_90 = df.query('수학>=90')
print(morethan_90)
```

```python
#Boxoffice.xlsx 파일을 데이터프레임으로 생성하기
Boxdf = pd.read_excel('BoxOffice.xlsx')
Boxdf
#데이터프레임의 정보를 출력
Boxdf.info()
#대표국적별 영화 편수를 집계
value_counts = Boxdf['국적'].value_counts()
print(value_counts)
#대표국적이 '영국'인 데이터 출력
England = Boxdf[Boxdf['대표국적'] == '영국']
print(England)
#대표국적별 최상위 순위를 출력
highest = Boxdf.groupby('순위').min()
print(highest)
```
